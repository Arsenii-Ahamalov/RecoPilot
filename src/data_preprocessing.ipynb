```python
import pandas as pd
import numpy as np
movies = pd.read_csv('../data/movies.dat', sep='::', header=None, names=['movieId', 'title', 'genres'],
encoding='ISO-8859-1',engine='python')
ratings = pd.read_csv('../data/ratings.dat', sep='::', header=None, names=['userId', 'movieId', 'rating', 'timestamp'],
encoding='ISO-8859-1',engine='python')
users = pd.read_csv('../data/users.dat', sep='::', header=None,
 names=['userId', 'gender', 'age', 'occupation', 'zip-code'],encoding='ISO-8859-1',engine='python')
```

```python
movies.info()
ratings.info()
users.info()
```

```python
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
ratings['timestamp'] = ratings['timestamp'].dt.date
```

```python
movies['release_year'] = movies['title'].str.extract(r'\((\d{4})\)')
movies['release_year'] = movies['release_year'].astype(int)
```

```python
tmp = pd.merge(movies,ratings,on = 'movieId')[['timestamp','release_year']]
tmp['year'] = pd.to_datetime(tmp['timestamp']).dt.year
ratings['Age'] = tmp['year'] - tmp['release_year']
```

```python
movies.info()
ratings.info()
users.info()
```

```python
genre_dummies = movies['genres'].str.get_dummies(sep='|')
movies = pd.merge(movies,genre_dummies,left_index=True,right_index=True)
movies.drop(columns=['genres','title'],inplace=True)
movies.head()
```

```python
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'])
print(ratings.info())
print(ratings.head())
```

```python
users = pd.merge(users, users['gender'].str.get_dummies(),left_index=True,right_index=True)
users.drop(columns=['gender'],inplace=True)
users.head()
```

```python
users.drop(columns=['zip-code'],inplace=True)
users.head()
```
